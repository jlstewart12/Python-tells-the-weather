# WeatherPy

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

from api_keys import weather_api_key

from citipy import citipy

output_data_file = "output_data/cities.csv"

lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Cities List

```python
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)
```

### API Calls
* Performing a weather check on each city using a series of successive API calls.
* Included is a print log of each city as it's being processed (with the city number and city name).


```python
import json
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=" + units
```

```python
weather_response = requests.get(query_url)
weather_json = weather_response.json()

print(json.dumps(weather_json, indent=4, sort_keys=True))
```

```python
print(weather_json['weather'][0]['description'])
```

```python
name = []
clouds = []
country = []
date = []
humidity = []
lat = []
long = []
max_temp = []
wind_speed = []

for city in cities:
    query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=" + units
    response = requests.get(query_url).json()
    if response['cod'] == 200:
        weather_json = weather_response.json()
        name.append(response['name'])
        clouds.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        lat.append(response['coord']['lat'])
        long.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        wind_speed.append(response['wind']['speed'])
```

### Converting Raw Data into a DataFrame
* Exporting the city data into a .csv.
* Displaying the DataFrame

```python
city_data_pd = pd.DataFrame({'City': name, 'Cloudiness': clouds, 'Country': country, 'Date': date, 'Humidity': humidity, 'Lat': lat, 'Lng': long, 'Max Temp': max_temp, 'Wind Speed': wind_speed})
city_data_pd
```

```python
city_data_pd.to_csv("../output_data/cities.csv",index=True)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = northern.iloc[:,5]
y_values = northern.iloc[:,7]
plt.scatter(x_values,y_values)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.xlabel('Latitude')
plt.xlim(-5,90)
plt.ylim(-50,110)
plt.ylabel('Max Temp')
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(10,-40),fontsize=15,color="red")
print(f"The r-squared is: {rvalue}")
plt.savefig('../output_data/fig9.png')
plt.show()
```

# Analysis

## Trend 1: Max Temperature

#### Temperatures peak close to the equator and then decline as they approach either pole at 90 and -90 degrees

## Trend 2: Humidity

#### There is a slow rise in the humidity percantage in cities as the distance from the equator increases.

## Trend 3: North vs. South

#### The rates in which both temperatures drop and humidity rises are more intense in the Northern Hemisphere than the Southern Hemisphere
